```python
from gensim.models import Word2Vec
import multiprocessing
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('../sentiment140_tokenized.pickle')
```

```python
data.info()
```

```python
sentences = list(data['tokenized_text'].values)
```

```python
print(sentences[:3])
```

```python
len(sentences)
```

Source: https://blog.cambridgespark.com/tutorial-build-your-own-embedding-and-use-it-in-a-neural-network-e9cde4a81296

```python
EMBED_DIM = 300
emb = Word2Vec(sentences, size=EMBED_DIM, window=3, 
               min_count=3, negative=15, iter=1, workers=multiprocessing.cpu_count())
# get the word vector
word_vec = emb.wv
```

```python
dir(word_vec)
```

```python
# an example of each vocab and its weight?
print(word_vec.vectors[0])
```

```python
word ='cry'
res = word_vec.similar_by_word(word)
print(f'Most similar to {word} is:', res[:5])
```
